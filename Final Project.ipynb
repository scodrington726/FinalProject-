## Introduction

### Racial Disparities in Maternal Health
##### Racial disparities in maternal mortality rates represent one of the most urgent public health crises in the United States. Despite advances in medical technology and maternal care, women of color, particularly Black and Indigenous women, continue to die from pregnancy-related complications at disproportionately higher rates than their White counterparts. These disparities reflect the complex interplay of systemic racism, socioeconomic inequality, implicit bias in healthcare, and barriers to quality prenatal and postpartum care. Understanding how race influences maternal health outcomes is essential not only for diagnosing the structural causes of these injustices but also for shaping equitable healthcare policies and practices that protect all mothers.

## Literature Review

Maternal mortality remains a serious public health issue in the United States, and the burden is not evenly distributed across racial and ethnic groups. National data show that Black women experience much higher maternal mortality rates than White, Hispanic, and Asian women. In 2023, the maternal mortality rate for Black women was 50.3 deaths per 100,000 live births, compared with 14.5 for White women, 12.4 for Hispanic women, and 10.7 for Asian women ([Hoyert](https://www.cdc.gov/nchs/data/hestat/maternal-mortality/2023/maternal-mortality-rates-2023.htm)). These differences show that maternal mortality is not only a clinical issue, but also a racial equity issue. Because maternal deaths often reflect failures in prevention, access, and quality of care, racial differences in maternal mortality are important indicators of broader inequities in the healthcare system.

Research also shows that Indigenous women experience disproportionately high maternal mortality rates. A JAMA study examining state-level maternal mortality trends from 1999 to 2019 found that maternal mortality rates were higher among American Indian and Alaska Native and Black populations than among Asian, Native Hawaiian, Pacific Islander, Hispanic, and White populations in most states ([Fleszar et al.](https://jamanetwork.com/journals/jama/fullarticle/2806661)). The study also found that the American Indian and Alaska Native population experienced the largest increase in median state maternal mortality rates over the study period, while the Black population had the highest median state maternal mortality rate overall ([Fleszar et al.](https://jamanetwork.com/journals/jama/fullarticle/2806661)). These findings are important because they show that racial disparities in maternal mortality are not limited to Black-White comparisons. Indigenous maternal health outcomes must also be examined because they reflect the effects of historical trauma, underfunded healthcare systems, geographic barriers, and unequal access to culturally appropriate care.

A substantial body of literature identifies structural racism as a central contributor to these inequities. Sealy-Jefferson et al. argue that many studies of racial inequities in maternal health focus too heavily on individual-level risk factors while paying less attention to structural racism, including policies, institutional practices, and social norms that reproduce racial inequality ([Sealy-Jefferson et al.](https://www.frontiersin.org/journals/public-health/articles/10.3389/fpubh.2022.860850/full)). This perspective is important because it shifts the explanation away from individual behavior and toward the social and institutional conditions that shape health outcomes. Similarly, the American College of Obstetricians and Gynecologists states that racism, not race itself, drives health inequities in obstetric and gynecologic care ([American College of Obstetricians and Gynecologists](https://www.acog.org/clinical-information/policy-and-position-statements/statements-of-policy/2025/racism-in-obstetrics-gynecology)). This distinction matters because race is a social category rather than a biological cause of poor maternal outcomes. Therefore, racial disparities in maternal mortality should be understood as the result of unequal treatment, unequal resources, and unequal exposure to health risks.

The literature also emphasizes the importance of severe maternal morbidity, which refers to serious complications during pregnancy, delivery, or postpartum that can have major short- and long-term health consequences. Jeffers et al. explain that severe maternal morbidity has increased in the United States and that Black birthing people experience a disproportionate share of these complications. Their study found that racialized economic segregation was associated with higher odds of severe maternal morbidity among Black birthing people, even after adjusting for individual and hospital factors ([Jeffers et al.](https://pubmed.ncbi.nlm.nih.gov/36542927/)). This is important because many maternal deaths are preceded by severe complications such as hemorrhage, hypertensive disorders, cardiovascular problems, sepsis, or other pregnancy-related conditions. Studying morbidity alongside mortality helps show how racial inequities develop before death occurs and why timely intervention is essential.

Recent research also confirms that Black and Indigenous women face higher risks of pregnancy-related death. A JAMA Network Open study of pregnancy-related deaths from 2018 to 2022 found that American Indian and Alaska Native women had the highest pregnancy-related mortality rate, followed by non-Hispanic Black women ([Chen et al.](https://jamanetwork.com/journals/jamanetworkopen/fullarticle/2832320)). In that study, American Indian and Alaska Native women had a mortality rate 3.8 times higher than White women, while non-Hispanic Black women had a rate 2.8 times higher than White women ([Chen et al.](https://jamanetwork.com/journals/jamanetworkopen/fullarticle/2832320)). These findings strongly support the focus of this study because they show that both Black and Indigenous women experience substantially worse maternal health outcomes than White women.

Geographic and institutional factors further contribute to racial disparities in maternal mortality. Fleszar et al. found that maternal mortality rates varied significantly by state and that disparities for Black and American Indian and Alaska Native populations persisted across the study period ([Fleszar et al.](https://jamanetwork.com/journals/jama/fullarticle/2806661)). This suggests that place, policy, and healthcare infrastructure all influence maternal survival. Women in underserved or rural areas may face limited access to obstetric services, emergency care, transportation, and postpartum follow-up. At the institutional level, hospitals that serve higher proportions of Black and Indigenous patients may have fewer resources, which can worsen outcomes for patients already facing systemic barriers.

Overall, the existing literature makes clear that race is a critical factor in maternal mortality, but race itself is not the biological cause of these differences. Instead, the evidence points to structural racism, provider bias, unequal access to care, geographic barriers, and institutional inequities as major drivers of poor maternal outcomes for Black and Indigenous women. Although previous research has established that Black and Indigenous women experience disproportionately high maternal mortality rates, this study uses provisional maternal mortality data to compare maternal mortality rates across racial groups and test whether the differences between Black, Indigenous, and White women are statistically significant.

## Methods

### Study Design
This study uses an observational, cross-sectional design to examine racial differences in maternal mortality outcomes in the United States. Because the data are publicly available and organized by race and ethnicity, this study does not involve direct participant recruitment or collection of personal identifying information.

### Data Source
The dataset used in this analysis is the VSRR Provisional Maternal Death Counts and Rates dataset. The dataset includes provisional maternal death counts and maternal mortality rates by race and ethnicity.

### Variables
The main predictor variable is race/ethnicity. The main outcome variables are maternal death counts and maternal mortality rates.

Race/ethnicity was treated as a categorical variable. Maternal death count and maternal mortality rate were treated as continuous variables.

### Data Analysis
Data were analyzed using Python in VS Code. Descriptive statistics were calculated for the continuous variables, including the mean, median, mode, and range. Frequencies and proportions were calculated for the categorical race/ethnicity variable.

Independent samples t-tests were used as the inferential statistical test. The first t-test compared maternal mortality rates between Black women and White women. The second t-test compared maternal mortality rates between Indigenous women and White women. A significance level of p < 0.05 was used to determine whether the differences were statistically significant.

### Inferential Statistics: T-Tests
##### Independent samples t-tests were conducted to determine whether maternal mortality rates differed significantly between selected racial groups. The first t-test compared Black women and White women. The t-test produced a t-statistic of 30.017889373440934 and a p-value of 2.980751729911721e-55. Since the p-value was less than 0.05, the difference between Black and White women was statistically significant. The second t-test compared Indigenous women and White women. The t-test produced a t-statistic of 26.677825616746585 and a p-value of 3.6115776416756034e-13. Since the p-value was less than 0.05, the difference between Indigenous and White women was statistically significant.

```python
black_rates = race_df[race_df[race_col].str.contains("Black", case=False, na=False)][rate_col].dropna()

white_rates = race_df[race_df[race_col].str.contains("White", case=False, na=False)][rate_col].dropna()

indigenous_rates = race_df[
    race_df[race_col].str.contains("American Indian|Alaska Native|Indigenous", case=False, na=False)
][rate_col].dropna()

print("Black group size:", len(black_rates))
print("White group size:", len(white_rates))
print("Indigenous group size:", len(indigenous_rates))
```

```python

black_white_ttest = stats.ttest_ind(
    black_rates,
    white_rates,
    equal_var=False,
    nan_policy="omit"
)

indigenous_white_ttest = stats.ttest_ind(
    indigenous_rates,
    white_rates,
    equal_var=False,
    nan_policy="omit"
)

print("Black vs. White")
print("T-statistic:", black_white_ttest.statistic)
print("P-value:", black_white_ttest.pvalue)

print("\nIndigenous vs. White")
print("T-statistic:", indigenous_white_ttest.statistic)
print("P-value:", indigenous_white_ttest.pvalue)
```

## Discussion

### Summary of Main Findings
This study examined how race influences maternal health outcomes in the United States. The findings showed that maternal mortality rates differed across racial groups, with Black and Indigenous women experiencing higher rates than White women. These findings support the hypothesis that Black and Indigenous women experience worse maternal health outcomes compared to White women.

### Comparison to Previous Research
These findings align with previous research showing that Black and Indigenous women experience disproportionately high rates of maternal mortality. The results suggest that racial disparities in maternal health outcomes are not only individual medical issues, but also reflect broader structural problems, including differences in healthcare access, quality of care, implicit bias, and social determinants of health.

### Public Health Implications
The findings means reducing racial disparities in maternal mortality requires public health interventions that improve access to prenatal and postpartum care, expand culturally competent care, and address implicit bias within healthcare systems. Policies should also focus on improving care for Black and Indigenous mothers before, during, and after pregnancy.

### Limitations
Several limitations should be considered when interpreting the results. First, this study uses secondary data, so the analysis is limited by the variables and categories already included in the dataset. Second, because this is a cross-sectional analysis, the results cannot prove that race directly causes differences in maternal mortality rates. Third, maternal mortality is influenced by many other factors, including income, geography, insurance status, chronic health conditions, and access to care, which may not be fully captured in this dataset. Finally, provisional data may be updated over time, so the results should be interpreted with caution.

## Conclusion

These findings highlight the importance of addressing structural racism, healthcare access, and quality of maternal care as public health priorities.
